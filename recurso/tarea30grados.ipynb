```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(x1, x2):
    """Funición a evaluar"""
    a = (np.e**(-x1**2 - x2**2))*x1
    return a


def grad(x1, x2):
    """ Retorna gradiente  ∇f"""
    list1 = [1-2*x1**2, -2*x1*x2]
    descenso = np.array(list1)*math.e**(-x1**2 - x2**2)
    return descenso


def hessiano(x, y):
    """Evalua la matriz Hessiana, ingresar derivadas"""
    axax = (2*x*((2*x**2)-3))*math.e**(-(x**2) - (y**2))
    ayay = (2*x*((2*y**2) - 1))*math.e**(-(x**2) - (y**2))
    axay = 2*(2*x**2 - 1)*y*math.e**(-x**2 - y**2)
    # return axay
    return np.array([
        [axax, axay],
        [axay, ayay]
    ])

def dirgrad(x1, x2):
    """Retorna la dirección del gradiente p"""
    vgrad = grad(x1, x2)
    magGrad = np.sqrt(vgrad.dot(vgrad))
    p = -vgrad/magGrad
    return p

def phiAlpha(x0, x, d):
    paX1 = x0[0] + x * d[0]
    paX2 = x0[1] + x * d[1]
    return f(paX1, paX2)
```

```python
rad = math.radians(80)
print(rad)

p = np.array([
    np.cos(rad),
    np.sin(rad)
])

alpha = np.linspace(0, 2, 100)


def puntos(x0, alfa, p):
    x = x0 + p*alfa
    return f(x1=x[0], x2=x[1])


def gp(x0, alfa, p):
    x = x0 + p*alfa
    return np.dot(grad(x[0], x[1]), p)


def hp(x0, alfa, p):
    x = x0 + p*alfa
    return np.dot(np.dot(hessiano(x[0], x[1]), p), p)


ppts = []
gpts = []
hpts = []


for i in alpha:
    ppts.append(puntos(x0, i, p))
    gpts.append(gp(x0, i, p))
    hpts.append(hp(x0, i, p))


plt.plot(alpha, ppts, 'r', label="fx ")
plt.plot(alpha, gpts, 'g', label="gp")
plt.plot(alpha, hpts, 'b', label="Hp")
plt.legend()
plt.grid()
plt.show()
```

```python
def exhaustivoRefinado(p, xini, alpha=0, h=0.1, tol=1e-6):
    """Busqueda de minimo con metodo exhaustivo refinado. puedes cambiar el paso
    Retorna f(a) y alpha
    """
    k = 0
    while h > tol:
        while phiAlpha(xini, alpha+h, p) < phiAlpha(xini, alpha, p):
            alpha = alpha + h
            fnow = phiAlpha(xini, alpha, p)
            # print(k, h, fnow)
            k += 1
        alpha = alpha-h
        h = h / 10
    return alpha
```
