```python
import numpy as np
```

```python
def f(x):
    return (x[0]-2)**2 + (x[1]-2)**2


def g(x):
    return np.array([
        2*(x[0]-2),
        2*(x[1]-2),
    ])


def H(x):
    return np.array([
        [2, 0],
        [0, 2]
    ])
```

```python
# n = variables
# m = restricciones
def barreraLog(x0, A, b, mu, tol=1e-6):
    k = 0
    xk = x0
    T = 1e-1
    print("k, x, mu")
    while k < 1000:
        c, c1, c2 = calcC(A, xk, b)
        # print(f"c: {c}")
        grad = g(xk) - A.T.dot(c1)*mu
        # print(grad)
        hess = H(xk) + A.T.dot(c2).dot(A)*mu
        # print(hess)
        xk = xk - np.dot(grad, np.linalg.inv(hess))
        print(f"{k}, {xk}, {mu}")
        k = k + 1
        if (np.linalg.norm(grad) < T):
            mu /= 1.5
        if(mu < tol):
            break
    return xk
            


def calcC(A, x, b):
    m = A.shape[0]
    n = x.shape[0]
    C = A.dot(x)-b
    C1 = np.zeros(m)
    C2 = np.zeros((m, m))
    for i in range(0, m):
        val = 1/C[i]
        C1[i] = val
        C2[i, i] = val*val
    return C, C1, C2
```

```python
A = np.array([
    [1, 1],
    [2, -1]
])
b = np.array([4, 0])
x = np.array([2, 0])


barreraLog(x,A, b, mu = 100, tol=1e-20)
```
